# Bivariate Analysis

Now that we have analyzed each of the variables and its impact on the loan-status, let us take group of variables together and and analyze their combined effect on the loan-status. These categories are based on our business understanding. The original distribution column shows the average trend in all the data and we compare that with the data after applying our conditions.

```python
# Now let us look into variation of different categories for Loan_status using Bivariate Analysis in heatmap
def Bivariate_Categorical (*, data=df1, x, y, title=None, figsize=(8,6)):
    """
        Parameters:
            data : defaulted at the 'loan' variable, but can be changed externally if needed
            x, y : columns of data in the corresponding axis, both categorical
            figsize : a default have is given, but can be overridden
        Returns:
            None, does a heatmap plot
    """
    fig, ax = plt.subplots(figsize=figsize)
    pt = pd.pivot_table(data=df1, values='loan_status', index=y, columns=x)
    sns.heatmap(pt, ax=ax, cmap='YlGnBu', annot=True)
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
Bivariate_Categorical(x='home_ownership', y='loan_amnt')
```

Customers opting for very high loan amount irrespective of the type of home ownership have high chances of becomming defaulters.

```python
Bivariate_Categorical(x='term', y='purpose', figsize=(12,8))
```

As per the above plot, The customers opting for 60 months loan period term irrespective of all purposes are having more chances of defaulters.

```python
Bivariate_Categorical(x='addr_state', y=df['purpose'], figsize=(22,8))
```

As per the above plot, the darker the intersection of addr_state has with the purpose of the loan, the risker the loan application is. 

Some of the examples are below: 

•vacation loans in AK, HI, OR 

•education loans in AR, KS, UT 

•small business loans in DE, NM, WV, wY 

•wedding loans in MS, OR

```python
Bivariate_Categorical(x='annual_inc', y='loan_amnt', figsize=(22,8))
```

From the above plot, we can infer that the customers having very low income but asking very high loan amount are having higher chances of default rates.
